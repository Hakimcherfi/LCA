```python
def revert_3d_image(img):
    """3D flip"""
    return np.flip(img,2)

def revert_2d_image(img):
    """2D flip"""
    return np.flip(img,1)
```

```python
df = pd.read_csv('metadata.csv')
df = df.drop(columns = ['examId','seriesNo','roiX','roiY','roiZ','roiHeight','roiWidth','roiDepth'])

imagesList = []
labelsList = []
#2D processing, image selected = mean z
for pckfile in os.listdir("data"):
    with open("data/"+pckfile,"rb") as file:
        images = pck.load(file)
        ind = int(images.shape[0]/2)
        image = images[ind]
        imagesList.append(image/(np.max(image)))
        labelsList.append(float(df.loc[df.volumeFilename == pckfile,'aclDiagnosis']))
```
